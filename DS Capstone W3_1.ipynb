```python
import pandas as pd
```

```python
# Scrape the table from the following Wikipedia page
# df is list, df[0] is dataframe
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df = pd.read_html(url,header=0)
df[0]
```

```python
# Check dataframe column names
list(df[0].columns) 
```

```python
# Remove rows with Not assigned Borough
data = df[0]
new_data = data[data.Borough != 'Not assigned']
new_data
```

```python
# combine neighbourhood due to same borough
new_data1=new_data.groupby("Postcode").agg(lambda x:','.join(set(x)))
new_data2=new_data1.reset_index()
new_data2
```

```python
# check if there is any neighbourhood is not assigned but has borough 
new_data2[new_data2['Neighbourhood'].str.contains("Not assigned")]
```

```python
# assign the borough value to neighbourhood with not assigned
new_data2.loc[new_data2['Neighbourhood']=="Not assigned",'Neighbourhood']=new_data2.loc[new_data2['Neighbourhood']=="Not assigned",'Borough']
new_data2
```
